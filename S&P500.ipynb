```python
sp_df = sp_wiki_df_list[0]
```

```python
AppleData = yf.download('AAPL', period = '5y')
GoogleData = yf.download('GOOGL', period = '5y')
MicrosoftData = yf.download('MSFT', period = '5y')
IBMData = yf.download('IBM', period = '5y')
CiscoData = yf.download('CSCO', period = '5y')
```
